```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/social_media_vs_productivity.csv") 
df.head()
```

Here we check the number of rows and columns in the dataset (shape) and see what type of data each column contians (dtypes).

```python
df.shape, df.dtypes
```

info() shows an overview of all columns - their types, how many missing values they have and etc. 

```python
df.info()
```

This gives statistical summaries (mean, min, max, etc.) for all numerical columns. It helps us quickly see ranges and detect possible outliers. 

```python
df.describe()
```

Here we look at categorical (non-numeric) columns to see unique values and most frequent categories.

```python
df.describe(include="object")
```

We count how many missing values each column has.
This helps decide if we need to fill in missing data or remove rows later.

```python
df.isna().sum().sort_values(ascending=False)
```

To make visualizations work without errors, we temporarily fill missing numeric values with the column median.
This is not final cleaning â€” just for exploration.

```python
import numpy as np
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

A histogram shows the distribution of daily_social_media_time.
This helps us see how much time people typically spend on social media per day.

```python
sns.histplot(df["daily_social_media_time"], bins=30)
plt.title("Distribution: daily_social_media_time")
plt.xlabel("hours/day")
plt.show()
```

A boxplot helps visualize the spread of data and detect outliers (very high or low values).

```python
sns.boxplot(x=df["daily_social_media_time"])
plt.title("Boxplot: daily_social_media_time")
plt.show()
```

This scatter plot shows the relationship between daily_social_media_time and actual_productivity_score.
We can see if higher social media time tends to correlate with lower productivity.

```python
sns.scatterplot(x=df["daily_social_media_time"], y=df["actual_productivity_score"])
plt.title("Social media time vs. actual productivity")
plt.show()
```

A correlation heatmap shows how strongly numerical features are related.
Values near +1 mean strong positive correlation, and near -1 mean strong negative correlation.

```python
corr = df.select_dtypes("number").corr()
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=False)
plt.title("Correlation heatmap")
plt.show()
```
